```python
# prompt: download a stock with yfinance and show the percent increase or decrease of the current price of the stock from yesterday, from last week an from last 3 months in a colorful tabular table. also display it as chart in colors using plotly

import pandas as pd
import yfinance as yf
from plotly.express import line
import plotly.express as px

# Define the stock symbol and period.
# Define the start and end dates for the data you want to fetch.
stock_ticker = "AAPL"
start = "2024-01-01"
end = "2024-05-20"

# Download the XU100 (BIST 100) data from Yahoo Finance.
stock_data = yf.download(stock_ticker, start=start, end=end)

# Check if the stock_data DataFrame is empty.
if len(stock_data) > 0:
    # Calculate the percent change from yesterday, last week, and last 3 months.
    percent_change_yesterday = (stock_data["Close"][-1] - stock_data["Close"][-2]) / stock_data["Close"][-2] * 100
    percent_change_last_week = (stock_data["Close"][-1] - stock_data["Close"][-5]) / stock_data["Close"][-5] * 100
    percent_change_last_3_months = (stock_data["Close"][-1] - stock_data["Close"][-60]) / stock_data["Close"][-60] * 100
else:
    print("The stock_data DataFrame is empty. Cannot calculate percent changes.")

data = {
    "Period": ["Yesterday", "Last Week", "Last 3 Months"],
    stock_ticker: [percent_change_yesterday, percent_change_last_week, percent_change_last_3_months],
}

df = pd.DataFrame(data)

# Transpose the DataFrame to switch rows and columns.
df_transposed = df.transpose()

# Set the new column names.
df_transposed.columns = df_transposed.iloc[0]

# Drop the first row which contains the old column names.
df_transposed = df_transposed.iloc[1:]

# Display the transposed DataFrame.
display(df_transposed)
```

```python
# prompt: convert the above code so that it takes a list of stocks, calculate the percentage changes from yesterday, last week and last 3 months and display in table. Each stock and its percentages must be a row.

import pandas as pd
import yfinance as yf
from plotly.express import line
import plotly.express as px
from datetime import date


def calculate_percent_changes(stocks):
  """
  Calculates the percent changes from yesterday, last week, and last 3 months for a list of stocks.

  Args:
    stocks: A list of stock tickers.

  Returns:
    A Pandas DataFrame containing the percent changes for each stock.
  """

  data = []
  for stock_ticker in stocks:
    stock_data = yf.download(stock_ticker, start="2024-01-01", end=date.today())

    if len(stock_data) > 0:
      percent_change_yesterday = (stock_data["Close"][-1] - stock_data["Close"][-2]) / stock_data["Close"][-2] * 100
      percent_change_last_week = (stock_data["Close"][-1] - stock_data["Close"][-5]) / stock_data["Close"][-5] * 100
      percent_change_last_3_months = (stock_data["Close"][-1] - stock_data["Close"][-60]) / stock_data["Close"][-60] * 100
    else:
      print(f"The stock_data DataFrame for {stock_ticker} is empty. Cannot calculate percent changes.")
      continue

    data.append({
      "Stock": stock_ticker,
      "Yesterday": percent_change_yesterday,
      "Last Week": percent_change_last_week,
      "Last 3 Months": percent_change_last_3_months,
    })

  df = pd.DataFrame(data)
  return df


# Example usage
stocks = [
    "AKBNK.IS", "ARCLK.IS", "ASELS.IS", "BIMAS.IS", "BRISA.IS", "CEKMA.IS",
    "DOHOL.IS", "DSUYP.IS", "EKGYO.IS", "ENKAI.IS", "EREGL.IS", "FROTO.IS",
    "GARAN.IS", "GUBRF.IS", "HALKB.IS", "HEKTS.IS", "ISCTR.IS", "KCHOL.IS",
    "KOZAA.IS", "KOZAL.IS", "KRDMD.IS", "MGROS.IS", "NERED.IS", "PETKM.IS",
    "PGSUS.IS", "SAHOL.IS", "SISE.IS", "TAVHL.IS", "THYAO.IS", "TKFEN.IS",
    "TTKOM.IS", "TUPRS.IS", "ULKER.IS", "VESTL.IS", "YKBNK.IS", "ZOREN.IS"
]
df = calculate_percent_changes(stocks)

# Create a style function to color the cells based on the percent change.
def color_percent_change(val):
    if val > 0:
        color = 'green'
    elif val < 0:
        color = 'lightcoral'
    else:
        color = 'blue'
    return f'color: {color}'

columns_to_apply = ["Yesterday", "Last Week", "Last 3 Months"]
df_new = df.style.applymap(color_percent_change, subset=columns_to_apply)

# Display the DataFrame
display(df_new)
```

```python
import pandas as pd
import yfinance as yf
from datetime import date


def calculate_percent_changes(stocks):
  """
  Calculates the percent changes for various periods for a list of stocks.

  Args:
    stocks: A list of stock tickers.

  Returns:
    A Pandas DataFrame containing the percent changes for each stock.
  """

  data = []
  for stock_ticker in stocks:
    stock_data = yf.download(stock_ticker, start="2024-01-01", end=date.today())

    if len(stock_data) > 0:
      percent_change_yesterday = (stock_data["Close"][-1] - stock_data["Close"][-2]) / stock_data["Close"][-2] * 100
      percent_change_day_before_yesterday = (stock_data["Close"][-1] - stock_data["Close"][-3]) / stock_data["Close"][-3] * 100
      percent_change_last_week = (stock_data["Close"][-1] - stock_data["Close"][-5]) / stock_data["Close"][-5] * 100
      percent_change_last_month = (stock_data["Close"][-1] - stock_data["Close"][-21]) / stock_data["Close"][-21] * 100  # Assuming 21 trading days in a month
      percent_change_last_3_months = (stock_data["Close"][-1] - stock_data["Close"][-60]) / stock_data["Close"][-60] * 100
    else:
      print(f"The stock_data DataFrame for {stock_ticker} is empty. Cannot calculate percent changes.")
      continue

    data.append({
      "Stock": stock_ticker,
      "Yesterday": percent_change_yesterday,
      "Day Before Yesterday": percent_change_day_before_yesterday,
      "Last Week": percent_change_last_week,
      "Last Month": percent_change_last_month,
      "Last 3 Months": percent_change_last_3_months,
    })

  df = pd.DataFrame(data)
  return df


# Example usage
stocks = [
    "AKBNK.IS", "ARCLK.IS", "ASELS.IS", "BIMAS.IS", "BRISA.IS", "CEKMA.IS",
    "DOHOL.IS", "DSUYP.IS", "EKGYO.IS", "ENKAI.IS", "EREGL.IS", "FROTO.IS",
    "GARAN.IS", "GUBRF.IS", "HALKB.IS", "HEKTS.IS", "ISCTR.IS", "KCHOL.IS",
    "KOZAA.IS", "KOZAL.IS", "KRDMD.IS", "MGROS.IS", "NERED.IS", "PETKM.IS",
    "PGSUS.IS", "SAHOL.IS", "SISE.IS", "TAVHL.IS", "THYAO.IS", "TKFEN.IS",
    "TTKOM.IS", "TUPRS.IS", "ULKER.IS", "VESTL.IS", "YKBNK.IS", "ZOREN.IS"
]
df = calculate_percent_changes(stocks)

# Function to color cells based on percent change
def color_percent_change(val):
    if val > 0:
        color = 'green'
    elif val < 0:
        color = 'lightcoral'
    else:
        color = 'blue'
    return f'color: {color}'

# Apply color style to specific columns
columns_to_apply = ["Yesterday", "Day Before Yesterday", "Last Week", "Last Month", "Last 3 Months"]
df_styled = df.style.applymap(color_percent_change, subset=columns_to_apply)

# Display the DataFrame
display(df_styled)
```
